```python
import text2emotion as te
import tweepy
import numpy as np
from textblob import TextBlob
import pandas as pd
import matplotlib.pyplot as plt
from tkinter import *
import re
global df
df = pd.DataFrame()
twitteraccount = ""
noOfTweet = 0
pos = 0
neg = 0
neutrall = 0
happysum = 0
sadsum = 0
angersum = 0
surprisesum = 0
fearsum = 0
def generate_graph():
    global df
    plt.xlabel('DATES')
    plt.ylabel('LIKES')
    time_likes=pd.Series(data=df['Likes'].values,index=df['Date'])
    time_likes.plot(figsize=(16,4),color='r')
    plt.show()

def open():
    global top
    top=Toplevel()
    top.config(background="#009999")
    top.geometry("1000x1000")
    top.title("SENTIMENT_ANALYSIS")
    global percent_pos_tweets_label
    global percent_neg_tweets_label
    global percent_neu_tweets_label
    global get_location
    global get_followers
    global get_positive
    global get_neg
    global get_neutrall
    global perc_of_pos
    global perc_of_neg
    global perc_of_neutrall
    global get_happiness
    global get_sadness
    global get_anger
    global get_surprise
    global get_fear

    location_label=Label(top,text="USER LOCATION:",font=('Helvetica bold', 15),fg="white",bg="black",height=1,width=22)
    followers_label=Label(top,text="USER FOLLOWERS:",font=('Helvetica bold', 15),fg="white",bg="black",height=1,width=22)
    positive_tweets_label=Label(top,text="POSITIVE TWEETS:",font=('Helvetica bold', 15),fg="white",bg="black",height=1,width=22)
    negative_tweets_label=Label(top,text="NEGATIVE TWEETS:",font=('Helvetica bold', 15),fg="white",bg="black",height=1,width=22)
    neutrall_tweets_label=Label(top,text="NEUTRALL TWEETS:",font=('Helvetica bold', 15),fg="white",bg="black",height=1,width=22)
    percent_pos_tweets_label=Label(top,text="% OF POSITIVE TWEETS:",font=('Helvetica bold', 15),fg="white",bg="black",height=1,width=22)
    percent_neg_tweets_label=Label(top,text="% OF NEGATIVE TWEETS:",font=('Helvetica bold', 15),fg="white",bg="black",height=1,width=22)
    percent_neu_tweets_label=Label(top,text="% OF NEUTRALL TWEETS:",font=('Helvetica bold', 15),fg="white",bg="black",height=1,width=22)
    percent_happiness = Label(top, text="% OF HAPPINESS",font=('Helvetica bold', 15),fg="white",bg="black",height=1,width=22)
    percent_sadness = Label(top, text="% OF SADNESS",font=('Helvetica bold', 15),fg="white",bg="black",height=1,width=22)
    percent_anger = Label(top, text="% OF ANGER",font=('Helvetica bold', 15),fg="white",bg="black",height=1,width=22)
    percent_surprise = Label(top, text="% OF SURPRISE",font=('Helvetica bold', 15),fg="white",bg="black",height=1,width=22)
    percent_fear = Label(top, text="% OF FEAR",font=('Helvetica bold', 15),fg="white",bg="black",height=1,width=22)


    get_location=Entry(top,font=('Helvetica bold', 15),fg="white",bg="black",width=30)
    get_followers=Entry(top,font=('Helvetica bold', 15),fg="white",bg="black",width=30)
    get_positive=Entry(top,font=('Helvetica bold', 15),fg="white",bg="black",width=17)
    get_neg=Entry(top,font=('Helvetica bold', 15),fg="white",bg="black",width=17)
    get_neutrall=Entry(top,font=('Helvetica bold', 15),fg="white",bg="black",width=17)
    perc_of_pos=Entry(top,font=('Helvetica bold', 15),fg="white",bg="black",width=17)
    perc_of_neg=Entry(top,font=('Helvetica bold', 15),fg="white",bg="black",width=17)
    perc_of_neutrall=Entry(top,font=('Helvetica bold', 15),fg="white",bg="black",width=17)
    get_happiness=Entry(top,font=('Helvetica bold', 15),fg="white",bg="black",width=17)
    get_fear=Entry(top,font=('Helvetica bold', 15),fg="white",bg="black",width=17)
    get_anger=Entry(top,font=('Helvetica bold', 15),fg="white",bg="black",width=17)
    get_sadness=Entry(top,font=('Helvetica bold', 15),fg="white",bg="black",width=17)
    get_surprise=Entry(top,font=('Helvetica bold', 15),fg="white",bg="black",width=17)


    pie_on_polarity=Button(top,text="PIE ON POLARITY",font=('Helvetica bold', 15),fg="black",bg="#0099ff",height=2,width=20,command=generate_pie_1)
    pie_on_emotions=Button(top,text="PIE ON EMOTIONS",font=('Helvetica bold', 15),fg="black",bg="#0099ff",height=2,width=20,command=generate_pie_2)
    clear_data=Button(top,text="CLEAR",font=('Helvetica bold', 15),fg="black",bg="#0099ff",command=clear_all,height=2,width=20)
    generate=Button(top,text="GENERATE",font=('Helvetica bold', 15),fg="black",bg="#0099ff",command=Sentiment_Analysis,height=2,width=20)
    graph_on_tweet=Button(top,text="LIKE GRAPH",font=('Helvetica bold', 15),fg="black",bg="#0099ff",command=generate_graph,height=2,width=20)
    location_label.grid(row=0,column=1,padx=3,pady=2)
    get_location.grid(row=0,column=2,padx=3,pady=2)
    followers_label.grid(row=1,column=1,padx=3,pady=2)
    get_followers.grid(row=1,column=2,padx=3,pady=2)

    percent_happiness.grid(row=8,column=0,padx=3,pady=5)
    get_happiness.grid(row=8,column=1,padx=3,pady=5)
    percent_fear.grid(row=9,column=0,padx=3,pady=5)
    get_fear.grid(row=9,column=1,padx=3,pady=5)
    percent_anger.grid(row=10,column=0,padx=3,pady=5)
    get_anger.grid(row=10,column=1,padx=3,pady=5)
    percent_sadness.grid(row=11,column=0,padx=3,pady=5)
    get_sadness.grid(row=11,column=1,padx=3,pady=5)
    percent_surprise.grid(row=12,column=0,padx=3,pady=5)
    get_surprise.grid(row=12,column=1,padx=3,pady=5)


    positive_tweets_label.grid(row=2,column=0,padx=3,pady=5)
    get_positive.grid(row=2,column=1,padx=3,pady=5)
    negative_tweets_label.grid(row=3,column=0,padx=3,pady=5)
    get_neg.grid(row=3,column=1,padx=3,pady=5)
    neutrall_tweets_label.grid(row=4,column=0,padx=3,pady=5)
    get_neutrall.grid(row=4,column=1,padx=3,pady=5)

    percent_pos_tweets_label.grid(row=5,column=0,padx=3,pady=5)
    perc_of_pos.grid(row=5,column=1,padx=3,pady=5)
    percent_neg_tweets_label.grid(row=6,column=0,padx=3,pady=5)
    perc_of_neg.grid(row=6,column=1,padx=3,pady=5)
    percent_neu_tweets_label.grid(row=7,column=0,padx=3,pady=5)
    perc_of_neutrall.grid(row=7,column=1,padx=3,pady=5)
    generate.grid(row=13, column=0, padx=20, pady=20)
    pie_on_polarity.grid(row=13,column=2,padx=20,pady=20)
    pie_on_emotions.grid(row=14,column=2,padx=20,pady=20)
    graph_on_tweet.grid(row=14,column=0,padx=20,pady=20)
    clear_data.grid(row=15,column=0,padx=20,pady=20)
def clear_all():
    entry_account.delete(0, END)
    entry_tweet.delete(0, END)
    entry_date_from.delete(0, END)
    entry_date_to.delete(0, END)
    get_location.delete(0, END)
    get_followers.delete(0, END)
    get_positive.delete(0,END)
    get_neg.delete(0,END)
    get_neutrall.delete(0,END)
    perc_of_pos.delete(0,END)
    perc_of_neg.delete(0,END)
    perc_of_neutrall.delete(0,END)
    get_happiness.delete(0,END)
    get_anger.delete(0,END)
    get_surprise.delete(0,END)
    get_fear.delete(0,END)
    get_sadness.delete(0,END)




def percentagecalculator(list1):
    sum = 0
    for x in list1:
        sum = sum + x
    return sum


def cleanUpTweets(text):
    text = re.sub(r'@[A-Za-z0-9_]+', '', text)
    text = re.sub(r'#', '', text)
    text = re.sub(r'RT : ', '', text)
    text = re.sub(r'https?:\/\/[A-Za-z0-9\.\/]+', '', text)
    return text


def gettextSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity


def gettextPolarity(text):
    return TextBlob(text).sentiment.polarity


def gettextanalysis(a):
    if a > 0:
        return 'Positive'
    elif a == 0:
        return 'Neutral'
    else:
        return 'Negative'


def get_english(message):
    analysis = TextBlob(message)
    language = analysis.detect_language()
    if language == 'en':
        return message
    return str(analysis.translate(to='en'))


def generate_pie_1():
    explode = (0, 0.1, 0)
    labels = 'Positive', 'Negative', 'Neutral'
    sizes = [pos, neg, neutrall]
    colors = ['yellowgreen', 'lightcoral', 'gold']
    plt.pie(sizes, explode=explode, colors=colors, autopct='%1.1f%%', startangle=120)
    plt.legend(labels, loc=(-0.05, 0.05), shadow=True)
    plt.axis('equal')
    plt.show()


def generate_pie_2():
    y = np.array([happysum, sadsum, angersum, surprisesum, fearsum])
    mylabels = ["HAPPY", "SAD", "ANGRY", "SURPRISE", "FEAR"]
    colors = ['yellowgreen', 'lightcoral', 'gold', '#00ffff', '#4000ff']
    myexplode = [0.1, 0, 0, 0, 0]
    plt.pie(y, labels=mylabels, explode=myexplode, colors=colors, autopct='%1.1f%%', startangle=120)
    plt.legend(mylabels, loc=(-0.05, 0.05), shadow=True)
    plt.axis('equal')
    plt.show()

def Sentiment_Analysis():
    apikey = '6nUAkdWvFyOVDTYqRRaEAaifp'
    apisecret = 'sn3r544SjAtnWcy8gEEdsxUUT4T1uiNdlIFOGmbipHYxB8h8pp'
    accesstoken = '781414615813070848-33qvkVnmjqNwV5rs5b7Xp66trIDjmC7'
    accesstokensecret = 'hCTu45olGBfB56bC4PppqEKPMtKCmGnnoUMIjcRt3vjEU'

    auth = tweepy.OAuthHandler(apikey, apisecret)
    auth.set_access_token(accesstoken, accesstokensecret)
    twitterapi = tweepy.API(auth)
    twitteraccount = entry_account.get()
    noOfTweet = int(entry_tweet.get())
    if (noOfTweet == 0):
        return
    from_date=entry_date_from.get()
    to_date=entry_date_to.get()


    if(len(from_date)==0 and len(to_date)==0):
     tweets = tweepy.Cursor(twitterapi.user_timeline, screen_name=twitteraccount, count=None,since=None,until=None,
                           since_id=None, max_id=None, trim_user=True, exclude_replies=True, contributor_details=False,
                           include_entities=False).items(noOfTweet)

    elif(len(from_date)==0):
        tweets = tweepy.Cursor(twitterapi.user_timeline, screen_name=twitteraccount, count=None, since=None, until=to_date,
                               since_id=None, max_id=None, trim_user=True, exclude_replies=True,
                               contributor_details=False,
                               include_entities=False).items(noOfTweet)

    elif(len(to_date)==0):
        tweets = tweepy.Cursor(twitterapi.user_timeline, screen_name=twitteraccount, count=None, since=from_date, until=None,
                               since_id=None, max_id=None, trim_user=True, exclude_replies=True,
                               contributor_details=False,
                               include_entities=False).items(noOfTweet)

    else:
        tweets = tweepy.Cursor(twitterapi.user_timeline, screen_name=twitteraccount, count=None, since=from_date, until=to_date,
                               since_id=None, max_id=None, trim_user=True, exclude_replies=True,
                               contributor_details=False,
                               include_entities=False).items(noOfTweet)


    user=twitterapi.get_user(twitteraccount)
    location_user=user.location
    if(len(location_user)==0):
        location_user=location_user+"NONE"
    followers=user.followers_count
    get_location.insert(0,location_user)
    get_followers.insert(0,followers)
    global df
    like = []
    dates = []
    all_tweets=[]
    for tweet in tweets:
        like.append(tweet.favorite_count)
        dates.append(tweet.created_at)
        all_tweets.append(tweet.text)
    df['Tweet']=all_tweets
    df['Date']=dates
    df['Likes']=like
    df['Tweet'] = df['Tweet'].apply(cleanUpTweets)
    df = df.drop(df[df['Tweet'] == ''].index)
    df['Tweet'] = df['Tweet'].apply(get_english)
    df['Subjectivity'] = df['Tweet'].apply(gettextSubjectivity)
    df['Polarity'] = df['Tweet'].apply(gettextPolarity)
    df['Score'] = df['Polarity'].apply(gettextanalysis)
    positive = df[df['Score'] == "Positive"]
    negative = df[df['Score'] == "Negative"]
    neutral = df[df['Score'] == "Neutral"]
    global pos
    pos = positive.shape[0] / df.shape[0] * 100
    pos=round(pos,2)
    global neg
    neg = negative.shape[0] / df.shape[0] * 100
    neg=round(neg,2)
    global neutrall
    neutrall = neutral.shape[0] / df.shape[0] * 100
    neutrall=round(neutrall,2)
    x = str(positive.shape[0])
    get_positive.insert(0, x)
    y = str(negative.shape[0])
    get_neg.insert(0, y)
    z = str(neutral.shape[0])
    get_neutrall.insert(0, z)
    perc_of_pos.insert(0, pos)
    perc_of_neg.insert(0, neg)
    perc_of_neutrall.insert(0, neutrall)
    lt = df['Tweet']
    Happy = []
    Sad = []
    Angry = []
    Surprise = []
    Fear = []
    for text in lt:
        dict = te.get_emotion(text)
        for key in dict:
            if (key == "Happy"):
                Happy.append(dict[key])
            elif (key == "Sad"):
                Sad.append(dict[key])
            elif (key == "Angry"):
                Angry.append(dict[key])
            elif (key == "Surprise"):
                Surprise.append(dict[key])
            else:
                Fear.append(dict[key])

    df['HAPPY'] = Happy
    df['SAD'] = Sad
    df['ANGRY'] = Angry
    df['SURPRISE'] = Surprise
    df['FEAR'] = Fear
    global happysum
    happysum = percentagecalculator(Happy)
    global sadsum
    sadsum = percentagecalculator(Sad)
    global angersum
    angersum = percentagecalculator(Angry)
    global surprisesum
    surprisesum = percentagecalculator(Surprise)
    global fearsum
    fearsum = percentagecalculator(Fear)
    total = 0
    totalsum = zip(Happy, Sad, Angry, Surprise, Fear)
    for x in totalsum:
        for y in x:
            total = total + y
    happypercent = (happysum / total) * 100
    happypercent=round(happypercent,2)
    sadpercent = (sadsum / total) * 100
    sadpercent=round(sadpercent,2)
    angerpercent = (angersum / total) * 100
    angerpercent=round(angerpercent,2)
    surprisepercent = (surprisesum / total) * 100
    surprisepercent=round(surprisepercent,2)
    fearpercent = (fearsum / total) * 100
    fearpercent=round(fearpercent,2)
    get_happiness.insert(0, str(happypercent))
    get_sadness.insert(0, str(sadpercent))
    get_anger.insert(0, str(angerpercent))
    get_surprise.insert(0, str(surprisepercent))
    get_fear.insert(0, str(fearpercent))
    # with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    #       print(df)

if __name__ == "__main__":
    gui = Tk()
    gui.config(background="#003e53")
    gui.geometry("600x300")
    gui.title("LOGIN_PAGE")
    title = Label(gui, text="LOGIN", fg="white", bg="#003e53", font=("bold", 15))
    account = Label(gui, text="@ USER NAME:", fg="white", bg="#003e53", font=("bold", 15))
    nooftweets = Label(gui, text="TWEETS:", fg="white", bg="#003e53", font=("bold", 15))
    from_date = Label(gui, text="FROM:", fg="white", bg="#003e53", font=("bold", 15))
    to_date = Label(gui, text="TO:", fg="white", bg="#003e53", font=("bold", 15))
    submit_button = Button(gui, text="SUBMIT", fg="white", bg="#00cc99", command=open, height=1, width=7)
    clear_button = Button(gui, text="CLEAR", fg="white", bg="#00cc99", command=clear_all, height=1, width=7)

    entry_account = Entry(gui, font=("bold", 15))
    entry_tweet = Entry(gui, font=("bold", 15))
    entry_date_from = Entry(gui, font=("bold", 15))
    entry_date_to = Entry(gui, font=("bold", 15))

    title.grid(row=0, column=1, pady=20)

    account.grid(row=1, column=0, padx=5, pady=5)
    entry_account.grid(row=1, column=1, padx=5, pady=5)
    nooftweets.grid(row=2, column=0, padx=5, pady=5)
    entry_tweet.grid(row=2, column=1, padx=5, pady=5)
    from_date.grid(row=3, column=0, padx=5, pady=5)
    entry_date_from.grid(row=3, column=1, padx=5, pady=5)
    to_date.grid(row=4, column=0, padx=5, pady=5)
    entry_date_to.grid(row=4, column=1, padx=5, pady=5)
    submit_button.grid(row=5, column=1, pady=5)
    clear_button.grid(row=6, column=1, pady=5)
    gui.mainloop()
```
